```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
df=pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv')
df.head(10)
```

```python
df_line=df.groupby(['Year'])['Automobile_Sales'].sum()
df_line
```

```python
fig=plt.figure(figsize=(12,6))
df_line.plot(x=df_line.index,y=df_line.values)
plt.title('Automobile Sales during Recession')
```

```python
df.head(5)
```

```python
df_rec=df[df['Recession']==1]
df_rec
```

```python
df_line1=df_rec.groupby(['Year','Vehicle_Type'])['Automobile_Sales'].mean().reset_index()
df_line1
```

```python
df_line1.set_index('Year')
```

```python
df.head(5)
```

```python
df_new=df_line1.groupby('Vehicle_Type')['Automobile_Sales'].sum()
df_new
```

```python
df_new.plot(x=df_new.index,y=df_new.values)
plt.title('Automobile Sales during Recession')
plt.xlabel('Vehicles')
plt.ylabel('Sales')
```

```python
df.head(10)
```

```python
df_bar=df.groupby('Recession')['Automobile_Sales'].mean()
df_bar.columns=['Recession','Automobile_Sales']
df_bar
```

```python
sns.barplot(data=df_bar,x=df_bar.index,y=df_bar.values,hue=df_bar.index)
plt.xlabel('Recession')
plt.ylabel('Automobile_sales')
plt.title('Average Automobile Sales during Recession and Non-Recession')
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])
plt.show()

```

```python
df_recc=df[df['Recession']==1]
df_recc1=df_recc.groupby('Year')['GDP'].mean().reset_index()
df_recc1
```

```python
df_non=df[df['Recession']==0]
df_non1=df_non.groupby('Year')['GDP'].mean().reset_index()
df_non1
```

```python
fig=plt.figure(figsize=(10,6))
ax1=fig.add_subplot(1,2,1)
ax1.plot(df_recc1['Year'],df_recc1['GDP'])
ax1.set_title('During recession')

ax2=fig.add_subplot(1,2,2)
ax2.plot(df_non1['Year'],df_non1['GDP'])
ax2.set_title('During Non Recession')

fig.suptitle('Two plots in a Figure')
fig.tight_layout()
```
